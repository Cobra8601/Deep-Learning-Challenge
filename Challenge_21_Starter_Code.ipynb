## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np


#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
# Copy data set before manipulations
application_df_copy = application_df
application_df_copy = application_df_copy.drop(['EIN', 'NAME'], axis=1)
application_df_copy
```

```python
# Confirm columns
application_df_copy.columns
```

```python
# Determine the number of unique values in each column.
application_df_copy.nunique()
```

```python
# Look at APPLICATION_TYPE value counts for binning
# Counts for each application type: go into df, grab list APPLICATION_TYPE
# and count values for each type
app_type_counts = application_df_copy['APPLICATION_TYPE'].value_counts()
app_type_counts
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name 'application_types_to_replace`
application_types_to_replace = list(app_type_counts[app_type_counts < 500].index)

# Replace in dataframe
# use for loop and go in the df & grab 'application type' to replace with other
for app in application_types_to_replace:
    application_df_copy['APPLICATION_TYPE'] = application_df_copy['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
# go in the df and give the count of each app type in its bin
application_df_copy['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
# go in the df and get the count of each classification
class_count = application_df_copy['CLASSIFICATION'].value_counts()
class_count
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
# go into the first count and get items where count is >1
class_count2 = class_count.loc[class_count > 1]
class_count2.head()
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
# go into class count and grab only those less than 1000
classifications_to_replace = list(class_count[class_count < 1000].index)

# Replace in dataframe
for cls in classifications_to_replace:
    application_df_copy['CLASSIFICATION'] = application_df_copy['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure binning was successful
application_df_copy['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
numeric_data = pd.get_dummies(application_df_copy)
numeric_data.tail()
```

```python
# Split our preprocessed data into our features and target arrays
# Copy the data first
numeric_data_copy = numeric_data.copy()

# Features
X = numeric_data_copy.drop("IS_SUCCESSFUL", axis=1)
feature_names = X.columns
X.head()

# Labels / Array
y = numeric_data_copy["IS_SUCCESSFUL"].values
y[:100]
```

```python
# Counts for y; 0 = No , 1 = Yes
# Confirm the counts
numeric_data["IS_SUCCESSFUL"].value_counts()
```

```python
# Confirm the X value dataframe
X.head()
```

```python
# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=86)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
number_input_features = len(X_train_scaled[0])
hidden_nodes_layer1 = 50
hidden_nodes_layer2 = 40
hidden_nodes_layer3 = 30

nn_model = tf.keras.models.Sequential()

# First hidden layer
nn_model.add(tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="relu")
)

# Second hidden layer
nn_model.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="relu")
)

# Third hidden layer
nn_model.add(tf.keras.layers.Dense(units=hidden_nodes_layer3, activation="relu")
)

# Output layer
nn_model.add(tf.keras.layers.Dense(units=1, activation="relu"))

# Check the structure of the model
nn_model.summary()
```

```python
# Compile the model
nn_model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Train the model
fit_model = nn_model.fit(X_train_scaled, y_train, epochs=15)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn_model.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
